````python
# 1. Convert Height to inches
df[['feet', 'inches']] = df['Height'].str.extract(r'(\d+)\'(\d+)"').astype(int)


# 2. Remove "lbs" from Weight
df['Weight'] = df['Weight'].str.replace('lbs', '').astype(int)

# 3. Extract Team and Contract details
df["Team & Contract"] = df["Team & Contract"].str.replace('\n', '')
df[['Team', 'Contract']] = df['Team & Contract'].str.extract(r'(\D+)(\d+\s*~\s*\d+)')

# 4. Remove unnecessary whitespace
df['Team'] = df['Team'].str.strip()
df['Contract'] = df['Contract'].str.strip()

# 5. Remove Euro symbol and convert Value, Wage, and Release Clause to numeric
df['Value'] = df['Value'].str.replace('€', '').str.replace('M', '').str.replace('K', '').astype(float)
df ['Wage'] = df['Wage'].str.replace('€', '').str.replace('K', '').astype(float)
df['Release Clause'] = df['Release Clause'].str.replace('€', '').str.replace('M', '').str.replace('K', '').astype(float)
df["Hits"] = df["Hits"].str.replace('\n', '').str.replace('K','').astype(float)

# 6. Removing stars from the rest and transforming them to numbers
df["W/F"] = df["W/F"].str.replace("★", '').astype(int)
df['SM'] = df['SM'].str.replace("★", '').astype(int)
df['IR'] = df['IR'].str.replace("★", '').astype(int)

# Drop the original 'Team & Contract' column
df["height"] = df["feet"] * 30.48 + df["inches"] * 2.54
df["weight"] = df["Weight"] / 2.205
df = df.drop(['Team & Contract','Height', 'feet', 'inches', 'Weight'], axis=1)
````
